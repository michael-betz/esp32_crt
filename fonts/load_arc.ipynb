```python
glyphs = []
g_len = 0
g_lens = []
for ll in dat.splitlines():
    l = ll.split()
    if len(l) == 1:
        # add glyph width at the end        
        glyphs.append(int(l[0]) & 0x7F)
        g_len += 1
        g_lens.append(g_len)
        g_len = 0
        continue
    dats = [int(d) for d in l]
    t, a_x, a_y, b_x, b_y, fo, lo = dats
    # we encode as such:
    # 0: <0 :8> for line or <fo:4><lo:4> for arc
    # 1: a_x ...
    if t == 1:
        a = 0
    elif t == 2:
        a = fo << 4 | lo
    else:
        print("HUH!!")
    dats_ = [a, a_x, a_y, b_x, b_y]
    glyphs += dats_
    g_len += len(dats_)

g_inds = cumsum(g_lens)

savez('arc_data.npz', glyphs=glyphs, g_inds=g_inds)
```
