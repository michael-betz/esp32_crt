# Convert the arc font format

```python
# all glyphs
all_bs = bytes()
glyph_props = []

# one glyph as bytes blob
bs = bytes()
ce = CoordinateEncoder()
bb_x = MinMax()
bb_y = MinMax()
start_index = 0

for ll in dat.splitlines():
    l = ll.split()
    
    # glyph width is a line with one number only. It's the end of the glyph
    if len(l) == 1:
        # End marker not needed, will use end_index
        # bs += (T_END << 4).to_bytes()
        all_bs += bs
        glyph_props.append({
#             "start_index": start_index,
            "end_index": start_index + len(bs),
            "adv_w": (int(l[0]) & 0x7F) + 3,
#             "box_w": bb_x.get_max() - bb_x.get_min(),
#             "box_h": bb_y.get_max() - bb_y.get_min(),
        })
        
        start_index += len(bs)
        
        # reset internal states for the next glyph
        bs = bytes()
        ce = CoordinateEncoder()
        bb_x.reset()
        bb_y.reset()
        continue
    
    dats = [int(d) for d in l]
    t, a_x, a_y, b_x, b_y, fo, lo = dats
    
    # t indicates the type
    if t == 1:  # line
        bs += ce.encode([a_x, a_y], T_GOTO << 4)
        bb_x.add(a_x)
        bb_y.add(a_y)
        bs += ce.encode([b_x, b_y], T_LINETO << 4)
        bb_x.add(b_x)
        bb_y.add(b_y)
    elif t == 2:  # arc
        bs += ce.encode([a_x, a_y], T_ARC << 4)
        bs += b_x.to_bytes()
        bs += b_y.to_bytes()
        bs += (fo << 4 | lo).to_bytes()
        bb_x.add(a_x - b_x)
        bb_x.add(a_x + b_x)
        bb_y.add(a_y - b_y)
        bb_y.add(a_y + b_y)
    else:
        print("HUH!!")
```
