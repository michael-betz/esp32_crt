```python
x_, y_ = 0, 0

def d(x, y, mult=0.25 / 4):
  global x_, y_
  plot(x_, y_, 'or')
  distx = x - x_
  disty = y - y_
  dist = sqrt(distx * distx + disty * disty)
  print(f"line from {x_:3.1f}, {y_:3.1f} to {x:3.1f}, {y:3.1f}, dist: {dist:.3f}");

  n = dist * mult
  n = int(n + 0.5)
  if n <= 0:
    return
  dx = distx / n
  dy = disty / n
  for i in range(n):
    x_ += dx
    y_ += dy
    plot(x_, y_, '.k')
    print(f'{x_:3.1f}, {y_:3.1f}')

d(0, 0)
axis('equal')
```

```python
d(-1000,-1000,  0)
d(-1000, 1000)
d( 1000, 1000)
d( 1000,-1000)
d(-1000,-1000)
d( -100, -100,  0)
d(  100,  100)
d( -100,  100,  0)
d(  100, -100)
```
