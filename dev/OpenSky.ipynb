```python
%pylab notebook
import requests
```

```python
url = 'https://opensky-network.org/api/states/all'

# use http://bboxfinder.com/#45.813486,5.822754,47.379754,8.250732
bbox = [5.822754,45.813486,8.250732,47.379754]
bbox_ = (bbox[3], bbox[1], bbox[2], bbox[0])  # for osmnx

data = {
    'lomin': bbox[0],
    'lamin': bbox[1],
    'lomax': bbox[2],
    'lamax': bbox[3]
}
```

```python
resp = requests.get(url, data)
print(resp)

dat = resp.json()
dat
states = dat['states']
```

```python
close()
plot()
for s in states:
    print(s[1], s[5], s[6])
    plot(s[5], s[6], 'o')
    
axis('equal')
tight_layout()
```

# Let's try OSMX

```python
import osmnx as ox
%matplotlib inline
ox.config(use_cache=True, log_console=True)
ox.__version__

center = "Geneva, Switzerland"
dist = 10 * 1000
```

```python
gdf = ox.features_from_address('Switzerland', dist=30000, tags={'admin_level': '4'})
```

```python
gdf.explore()
```

```python
# Find additional geometries
canons = ox.features_from_address(center, dist=dist, tags={"boundary": "administrative"})
water = ox.features_from_address(center, dist=dist, tags={"natural": ["water"]})
motoway = ox.features_from_address(
    center, dist=dist,
    tags={
        "highway": [
            "motorway",
            "motorway_link",
            "trunk",
            "trunk_link",
        ]
    },
)
```

```python
def print_names(gdb):
    for n in gdb['name']:
        if isinstance(n, str):
            print(n)
            
print_names(water)
print_names(motoway)
```

```python
fig, ax = subplots(1, 1)
ax.axis('equal')

gdf.plot(ax=ax)
water.plot(color="#466A8C", linewidth=1, ax=ax)
motoway.plot(color="#FF6ABC", linewidth=1, ax=ax)
```
