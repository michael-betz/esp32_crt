# Let's try OSMX

```python
import osmnx as ox
ox.config(use_cache=True, log_console=True)
ox.__version__

def print_names(gdb):
    for n in gdb['name']:
        if isinstance(n, str):
            print(n)

center = "Switzerland"
dist = 30 * 1000
```

```python
switzerland_border = ox.features_from_address(center, dist=dist, tags={'admin_level': '2'})
```

```python
water = ox.features_from_address(center, dist=dist * 4, tags={"natural": ["water"]})
water = water.assign(area=water['geometry'].area)
water = water.sort_values('area', ascending=False)
water.to_pickle('water.pkl')
```

```python
water = pd.read_pickle('water.pkl')
print_names(water)
```

```python
# drop lago maggiore and lago como (they are mostly outside switzerland anyway)
water = water.drop([('relation',   11758), ('relation',   541757)])
```

```python
fig, ax = subplots(1, 1)
ax.axis('equal')

switzerland_border.plot(ax=ax, color='grey')
water.head(5).plot(color="#466A8C", linewidth=1, ax=ax)
```

```python
def get_xy(gdf, simpl=0.01):
    poly = gdf['geometry'].simplify(simpl)
    return poly.exterior.xy
```

```python
gdf_all = pd.concat([switzerland_border, water.head(10)])

close()
for index, gdf in gdf_all.iterrows():
    x, y = get_xy(gdf, 0.005)
    plot(x, y, color='green')
    
for s in states:
    print(s[1], s[5], s[6])
    plot(s[5], s[6], 'o')
    
axis('equal')
axis(bbox)
tight_layout()
```
