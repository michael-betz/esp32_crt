```python
from fontTools.pens.boundsPen import BoundsPen
from fontTools.pens.basePen import BasePen
```

```python
gs = tt.getGlyphSet()
```

```python
bp = BoundsPen(gs)
gs["k"].draw(bp)
bp.bounds
```

```python
class MyPen(BasePen):
    def _lineTo(self, pt):
        print('lineTo', self.last_point, pt)
        self.last_point = pt
    
    def _moveTo(self, pt):
        print('moveTo', pt)
        self.last_point = pt
        
#     def _curveToOne(self, pt1, pt2, pt3):
#         print('curveToOne', pt1, pt2, pt3)
        
    def _qCurveToOne(self, pt1, pt2):
        pt0 = self.last_point
        print('qCurveToOne', pt0, pt1, pt2)
        self.last_point = pt2

mp = MyPen(gs)
```

```python
gs["k"].draw(mp)
```
